```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import sqlalchemy as sa
```

```python
# Configurar ruta relativa
ruta_datos = Path('..') / 'datos' / 'brutos' / 'ecommerce.db'

# Crear engine de SQLAlchemy
engine = sa.create_engine(f'sqlite:///{ruta_datos}')

# Verificar la conexión y tablas disponibles
inspector = sa.inspect(engine)
tablas = inspector.get_table_names()
print(f"Tablas disponibles: {tablas}")

# También puedes probar la conexión así:
with engine.connect() as conn:
    resultado = conn.execute(sa.text("SELECT name FROM sqlite_master WHERE type='table';"))
    print("Tablas:", resultado.fetchall())
```

```python
oct = pd.read_sql('2019-Oct', con=engine)
nov = pd.read_sql('2019-Nov', con=engine)
dic = pd.read_sql('2019-Dec', con=engine)
ene = pd.read_sql('2020-Jan', con=engine)
feb = pd.read_sql('2020-Feb', con=engine)
```
